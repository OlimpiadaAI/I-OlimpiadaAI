## Wstp

Klasyfikacja obrazw to proces przypisywania etykiety do obrazu na podstawie jego zawartoci. Przykadowo, chcielibymy, aby nasz program komputerowy mg rozpoznawa, czy na obrazie jest kot, pies, samochd, samolot czy moe co zupenie innego. W dzisiejszych czasach popularnym narzdziem do rozpoznawania obrazw s tzw. sieci konwolucyjne (CNN).

Sieci konwolucyjne s rodzajem sieci neuronowych, ktre potrafi analizowa i rozpoznawa wzorce w danych wizualnych. 

W przypadku klasyfikacji obrazw, sie konwolucyjna skada si z kilku warstw, w tym konwolucyjnych i poolingowych. Warstwy konwolucyjne su do ekstrakcji cech z obrazu, nastpnie za pomoc warstw poolingowych zmniejszamy wymiary danych, a na kocu wykorzystujemy warstwy w peni poczone do klasyfikacji obrazu.

Progresywne zmniejszanie warstw pozwala sieciom rozpoznawa coraz to bardziej abstrakcyjne cechy jako zoenie wielu pomniejszych cech np. ptak to co to ma dzib i jest opierzone. Dzib z kolei to np. ostry ksztat o tawym kolorze a opierzenie oznacza pokrycie du iloci maych kresek.

Twoimi danymi w tym zadaniu s obrazki w formacie \*.jpg o wymiarze 224 x 224. Obrazki te dziel si na dwie kategorie: *normal* oraz *onion*, ktrym przypisano odpowiednio etykiety 0 i 1.

Obrazki z klasy *normal* przedstawiaj jasnoszare figury na czarnym tle. Natomiast obrazki z klasy *onion* rni tym, e maj dodane ciemnoszare pasma tworzce warstwy w rodku jasnoszarych figur, co upodabnia je do cebuli. Wszystkie obrazki s dodatkowo zaszumione.




Publicznym intefrejsem klasy `YourCnnClassifier` musz by dwie metody ([class methods](https://stackoverflow.com/questions/12179271/meaning-of-classmethod-and-staticmethod-for-beginner) dokadnie rzecz biorc):

- `create_with_training` - ma wytrenowa model i zapisa jego parametry do pliku `cnn-classifier.pth`.

### Kryterium oceny

Twoje rozwizanie oceniane bdzie na podstawie skutecznoci klasyfikacji

$$
\mathrm{score}(accuracy) = \begin{cases}
    0 & \text{jeeli } accuracy < 0.5 \\
    (accuracy - 0.5) * 2 & \text{w.p.p.}
\end{cases}
$$

Powysze kryterium, klasa abstracyjna opisujca interfejs modelu oraz adowanie danych, s zaimplementowane poniej przez nas. Jednoczenie podany jest przykad trywialnego klasyfikatora, ktry zawsze twierdzi, e prbka jest normalna. Tym samym podczas testowania na zbalansowanym zbiorze testowym otrzymuje on 0 pkt.

```python
######################### NIE ZMIENIAJ TEJ KOMRKI PODCZAS WYSYANIA ##########################

import abc
from typing import Self
import os

import glob
import gdown
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
import torchvision
import matplotlib.pyplot as plt
import zipfile
```

## adowanie danych

```python
######################### NIE ZMIENIAJ TEJ KOMRKI PODCZAS WYSYANIA ##########################

GDRIVE_DATA = [
    ("1bR87z7ZI3gLK0vAGkyr_cnVGZ9P9bO7A", "train_data.zip"),
    ("1TA0lWnjJCv3lyRMML4JNHsJz3RJ-TUwZ", "valid_data.zip"),
]

def download_data():
    for file_id, zip_name in GDRIVE_DATA:
        folder_name = zip_name.split(".")[0]
        if not os.path.exists(folder_name):
            url = f'https://drive.google.com/uc?id={file_id}'
            gdown.download(url, output=zip_name, quiet=True)
            with zipfile.ZipFile(zip_name, 'r') as zip_ref:
                    zip_ref.extractall(folder_name)
            os.remove(zip_name)

download_data()

class ImageDataset(torch.utils.data.Dataset):
    """Implementacja abstrakcji zbioru danych z torch'a."""
    def __init__(self, dataset_type: str):
        self.filelist = glob.glob(f"{dataset_type}_data/*")
        self.labels   = [0 if "normal" in path else 1 for path in self.filelist]

    def __len__(self):
        return len(self.filelist)

    def __getitem__(self, idx) -> tuple[torch.Tensor, int]:
        if torch.is_tensor(idx):
            idx = idx.tolist()
        image = torchvision.transforms.functional.to_tensor(plt.imread(self.filelist[idx])[:,:,0])
        label = self.labels[idx]
        return image, label
    
    def loader(self, **kwargs) -> torch.utils.data.DataLoader:
        """
        Stwrz, `DataLoader`'a dla aktualnego zbioru danych.

        Wszystkie `**kwargs` zostan przekazane do konstruktora `torch.utils.data.DataLoader`.
        `DataLoader`'y w skrcie to abstrakcja adowania danych usdostpniajca wygodny interfejs.
        Moesz dowiedzie si o nich wicej tutaj: https://pytorch.org/docs/stable/data.html#torch.utils.data.DataLoader
        """
        return torch.utils.data.DataLoader(self, **kwargs)
    
train_dataset: ImageDataset = ImageDataset("train")
valid_dataset: ImageDataset = ImageDataset("valid")
```

## Kod z kryterium oceniajcym

```python
######################### NIE ZMIENIAJ TEJ KOMRKI PODCZAS WYSYANIA ##########################

def accuracy_to_points(accuracy: float) -> float:
    """Oblicz wynik na podstawie celnoci predykcji."""
    return (round(accuracy, 2) - 0.5) * 2 if accuracy > 0.5 else 0.0

def grade(model):
    """Oce ile punktw otrzyma aktualne zadanie."""
    model.eval()
    model

    test_loader = valid_dataset.loader()
    correct = 0
    total = 0
    with torch.no_grad():
        for [images, labels] in test_loader:
            outputs = model(images).squeeze()
            incorrect_indices = torch.where((outputs > 0.5).int() != labels)[0]
            correct += len(labels) - len(incorrect_indices)
            total += len(labels)
        accuracy = correct / total if total != 0 else 0
        if not FINAL_EVALUATION_MODE:
            print(f"Accuracy: {int(round(accuracy, 2) * 100)}%")
        return accuracy_to_points(accuracy)
```

## Publiczny interfejs rozwizania

Tylko tego wymagamy od Twojej klasy, w Twoim rozwizaniu moesz modyfikowa swoj klas do woli dodajc nowe metody oraz atrybuty klasy - cokolwiek co bdzie Ci potrzebne do rozwizania zadania.

```python
######################### NIE ZMIENIAJ TEJ KOMRKI PODCZAS WYSYANIA ##########################

class CnnClassifier(torch.nn.Module, abc.ABC):
    MODEL_PATH: str = "cnn-classifier.pth"
    
    @classmethod
    @abc.abstractmethod
    def load(cls) -> Self:
        """Zaaduj model z pliku."""
        
    
    @classmethod
    @abc.abstractmethod
    def create_with_training(cls) -> Self:
        """Zapisz model do pliku."""
        ...
```

## Przykadowe rozwizanie
Poniej prezentujemy proste rozwizanie, ktre w oczywisty sposb nie jest optymalne. Suy temu, aby byo wiadomo w jaki sposb ma dziaa cay notatnik.

```python
######################### NIE ZMIENIAJ TEJ KOMRKI PODCZAS WYSYANIA ##########################

if not FINAL_EVALUATION_MODE:
    class DummyCnnClassifier(CnnClassifier):
        def forward(self, x):
            batch_size, *_ = x.shape
            return torch.zeros(batch_size)

        @classmethod
        def load(cls) -> Self:
            return cls()
        
        @classmethod
        def create_with_training(cls) -> Self:
            return cls()
    
    dummy_model = DummyCnnClassifier.create_with_training()
    print(f"DummyCnnClassifier -- Ocena: {grade(dummy_model)} pkt")

    del dummy_model
    del DummyCnnClassifier
```

# Twoje Rozwizanie

```python
class YourCnnClassifier(CnnClassifier):
    def forward(self, x):
        batch_size, *_ = x.shape
        return torch.zeros(batch_size)

    @classmethod
    def load(cls) -> Self:
        return cls()
    
    @classmethod
    def create_with_training(cls) -> Self:
        return cls()
```

```python
######################### NIE ZMIENIAJ TEJ KOMRKI PODCZAS WYSYANIA ##########################

your_model = YourCnnClassifier.load() if FINAL_EVALUATION_MODE else YourCnnClassifier.create_with_training()
```

# Ewaluacja

```python
######################### NIE ZMIENIAJ TEJ KOMRKI PODCZAS WYSYANIA ##########################

def evaluate_model(model):
    """Oce ile punktw otrzyma aktualne zadanie."""
    return grade(model)
```

```python
######################### NIE ZMIENIAJ TEJ KOMRKI PODCZAS WYSYANIA ##########################

if not FINAL_EVALUATION_MODE:
    print(f"YourCnnClassifier -- Ocena: {evaluate_model(your_model):.2f} pkt")
```
