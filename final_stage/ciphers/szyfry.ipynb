Dzieci zaczy od przygotowania $n$ workw: po jednym dla kadego unikalnego znaku w caej bibliotece. Chopcy narysowali $3n$ rnych obrazkw, a dziewczynki wydrukoway naklejki z tymi obrazkami: kady obrazek zosta wydrukowany w pewnej nieznanej i bymoe rnej liczbie kopii. Naklejek jest jednak na tyle duo, e nie skoczsiw trakcie zabawy. Nastpnie, do kadego z przygotowanych workw dzieci wsypay naklejki odpowiadajce dokadnie trzem obrazkom. cznie wszystkie naklejki znalazy si w ktrymworku.

W czasie zabawy, dzieci postanowiy zaszyfrowa niektre wersy w ksikach. Dla napotkanego znaku, wycigay losow naklejk z waciwego dla danego znaku worka i naklejay j w jego miejsce. Dzieci, jak to dzieci, znudziy si, wic nie wszystkie wersy ksiek zostay przetworzone w ten sposb. Niestety obrazkw nie da si w aden sposb odklei, ale by moe dziki analizie ukadw literek w niezaklejonych tekstach bdziemy mogli odtworzy tekst z pozaklejanych wersw.

Napisz funkcj `decipher_corpus(clear_corpus, ciphered_corpus)`, ktra przyjmuje dwa argumenty:
* `clear_corpus`: lista zawierajca niezaszyfrowane stringi $-$ wersy, ktrych nie dotkny dzieci. Przykadowo:

```Python
clear_corpus = ["a tajn . u rzymian niewolnika , ktry ukrad , ", 
    "rodka : tam wszystko jasno owietlone . * i widz ", 
    "si do gry , utraci wasno * dc do centru ",
    "chciaabym widzie tego malca tylko na chwil . syszaem , ",
    "moci not , proszc o uwolnienie z tej usugi , "]
```

* `ciphered_corpus`: list z zaszyfrowanymi wersami $-$ wersy zaklejone naklejkami. Przykadowo:
```Python
ciphered_corpus = ["",
    "",
    ""]
```
Funkcja ta powinna rozszyfrowa list wersw `ciphered_corpus` zapisan szyfrem obrazkowym przy uyciu innych, niezaszyforwanych wersw `clear_corpus`. Funkcja powinna zwraca list rozszyfrowanych wersw.

## Ewaluacja

Kryterium oceny dla danego szyfru bdzie procent dobrze rozszyfrowanych znakw w tekcie (liczba prawidowo rozszyfrowanych znakw podzielona przez liczb wszystkich znakw w tekcie). Model osigajcy dokadno klasyfikacji poniej 50% otrzyma 0 punktw. 

Kocowy wynik bdzie redni z wynikw dla wszystkich testw i zostanie obliczony wedug wzoru:

$$\mathrm{score} =  \frac{1}{4} \sum_{i=1}^4 2 \cdot \max(\mathrm{accuracy}_i - 0.5, 0)$$
